```python
import pandas as pd
import numpy as np
import time
import requests
from bs4 import BeautifulSoup
```

```python
cleaned_members_csv = pd.read_csv('cleaned_members_export_25ea016a6a.csv')
subscribed_members_csv = pd.read_csv('subscribed_members_export_25ea016a6a.csv')
unsubscribed_members_csv = pd.read_csv('unsubscribed_members_export_25ea016a6a.csv')
```

```python
emails_cleaned = cleaned_members_csv['Email Address']
emails_subscribed = subscribed_members_csv['Email Address']
emails_unsubscribed = unsubscribed_members_csv['Email Address']
```

```python
# using dictionary to 
unique_emails = {}

for email in emails_cleaned:
    if email in unique_emails.keys():
        unique_emails[email] = unique_emails[email] + 1
    else:
        unique_emails[email] = 1

for email in emails_subscribed:
    if email in unique_emails.keys():
        unique_emails[email] = unique_emails[email] + 1
    else:
        unique_emails[email] = 1
        
for email in emails_unsubscribed:
    if email in unique_emails.keys():
        unique_emails[email] = unique_emails[email] + 1
    else:
        unique_emails[email] = 1
        
all_emails = pd.DataFrame(unique_emails.keys(), columns = ['email'])
all_emails['in_directory'] = 'nan' 
all_emails
```

```python
# scraping @berkeley.edu emails from the cal directory, stalling between requests to avoid 403 forbidden 
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
does_not_exist = 'No matches to your search. Please try again.'
base_URL = "https://www.berkeley.edu/directory/results"
indices = all_emails.index

for i in range(len(all_emails)):
    member = all_emails.iloc[i]
    query = '?search-term=' + member['email'].replace('@', '%40')
    full_query = base_URL + query
    page = requests.get(full_query, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    search_results = soup.find_all('section', {'class': 'search-results'})
    
    for result in search_results:
        if does_not_exist in result.text:
            all_emails.at[indices[i], 'in_directory'] = 0
            # member_activity['in_directory', indices[i]] = 0
        else:
            all_emails.at[indices[i], 'in_directory'] = 1
            # member_activity['in_directory', indices[i]] = 1
    
    print(str(member['email']) + " " + str(does_not_exist in result.text))
    time.sleep(5)
    
all_emails
```

```python
# filtering out emails that exist in the cal directory
graduated = all_emails['in_directory'] == 0
graduated_members = all_emails[graduated]
graduated_members.to_csv('graduated_email_list_sp21.csv')

graduated_members
```
